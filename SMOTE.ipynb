```python
from imblearn.over_sampling import SMOTE
import pandas as pd

df = pd.read_csv('combined_cough_sound_features.csv')
df.head()
```

```python
conditions_encoded, conditions_labels = df['Condition'].factorize()

X = df.drop(['ID', 'Condition'], axis=1)
y = conditions_encoded

healthy_count = (df['Condition'] == 'healthy').sum()

sampling_strategy = {0: healthy_count, 1: healthy_count}

smote = SMOTE(sampling_strategy=sampling_strategy)

X_resampled, y_resampled = smote.fit_resample(X, y)

resampled_df = pd.DataFrame(X_resampled, columns=X.columns)
resampled_df['Condition'] = conditions_labels[y_resampled]

resampled_df.to_csv('mark3_dataset.csv', index=False)
```

```python
condition_counts = resampled_df['Condition'].value_counts()

print(condition_counts)
```

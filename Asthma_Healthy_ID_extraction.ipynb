```python
def process_csv(file_path, healthy_ids, asthma_true_ids):

    df = pd.read_csv(file_path)


    print("Columns in DataFrame:", df.columns)

    healthy_ids.extend(df.loc[df['covid_status'] == 'healthy', 'id'].tolist())

    if 'asthma' in df.columns:

        asthma_mask = ((df['covid_status'] == 'healthy') | (df['asthma'].astype(str).str.upper() == 'TRUE')) & (df['asthma'].notna())
        asthma_true_ids.extend(df.loc[asthma_mask, 'id'].tolist())
    else:
        print("'asthma' column not found in DataFrame.")

colab_directory = '/content'

healthy_ids = []
asthma_true_ids = []

for filename in os.listdir(colab_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(colab_directory, filename)
        process_csv(file_path, healthy_ids, asthma_true_ids)

healthy_ids_set = set(healthy_ids)
asthma_true_ids_set = set(asthma_true_ids)

with open('healthy_ids2.txt', 'w') as f:
    f.write('\n'.join(map(str, healthy_ids_set)))

with open('asthma_true_ids2.txt', 'w') as f:
    f.write('\n'.join(map(str, asthma_true_ids_set)))
```
